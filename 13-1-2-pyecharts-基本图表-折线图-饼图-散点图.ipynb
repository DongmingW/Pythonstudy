### 折线图
线的类型(lineStyle)


```python
from pyecharts import Line
attr =["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
v1 =[5, 20, 36, 10, 10, 100]
v2 =[55, 60, 16, 20, 15, 80]
```

```python
# 普通折线图
line = Line('折线图')
line.add('商家A', attr, v1, mark_point=['max'])
line.add('商家B', attr, v2, mark_point=['min'], is_smooth=True)
#line.show_config()
#line.render('普通折线图.html')
line
```

```python
# 阶梯折线图
line2 = Line('阶梯折线图')
line2.add('商家A', attr, v1,  is_step=True, is_label_show=True)
#line2.show_config()
#line2.render('阶梯折线图.html')
line2
```

```python
# 面积折线图
line3 =Line("面积折线图")
line3.add("商家A", attr, v1,  line_opacity=0.2,   area_opacity=0.4, symbol=None, mark_point=['max'])
line3.add("商家B", attr, v2,  area_color='#a3aed5', area_opacity=0.3, is_smooth=True)
# line3.show_config()
#line3.render('面积折线图.html')
line3
```

## 饼图

```python
from pyecharts import Pie
attr =["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
v1 =[5, 20, 36, 10, 10, 100]
v2 =[55, 60, 16, 20, 15, 80]
```

```python
# 饼图
pie = Pie('饼图')
pie.add('芝麻饼', attr, v1, is_label_show=True)
# pie.show_config()
#pie.render('饼图.html')
pie
```

```python
# 玫瑰饼图
pie2 = Pie("饼图-玫瑰图示例", title_pos='center', width=900)
pie2.add("商品A", attr, v1, center=[25, 50], is_random=True, radius=[30, 75], rosetype='radius',is_label_show=True)
pie2.add("商品B", attr, v2, center=[75, 50], is_random=True, radius=[30, 75], rosetype='area', is_legend_show=False)
# pie2.show_config()
pie2.render('玫瑰饼图.html')
pie2
```

## 散点图

```python
from pyecharts import  Scatter
# 散点图
v1 =[10, 20, 30, 40, 50, 60]
v2 =[10, 20, 30, 40, 50, 60]
v3 =[60, 50, 40, 30, 20, 10]
scatter =Scatter("散点图示例")
scatter.add("A", v1, v2)
scatter.add("B", v1, v3)
#scatter.show_config()
#scatter.render('散点图.html')
scatter
```

```python
from pyecharts import  EffectScatter
attr =["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
v1 =[5, 20, 36, 10, 10, 100]
v2 =[55, 60, 16, 20, 15, 80]

# 动态散点图
es =EffectScatter("动态散点图")

# v1 x坐标 v2 y坐标
es.add("商家", v1, v2)
#es.show_config()
es.render('动态散点图.html')
es
```

```python
# 动态散点图各种图形
es = EffectScatter("动态散点图各种图形")
es.add("", [10], [10], symbol_size=20, effect_scale=3.5,  effect_period=3, symbol="pin")
es.add("", [20], [20], symbol_size=12, effect_scale=4.5, effect_period=4,symbol="rect")
es.add("", [30], [30], symbol_size=30, effect_scale=5.5, effect_period=5,symbol="roundRect")
es.add("", [40], [40], symbol_size=10, effect_scale=6.5, effect_brushtype='fill',symbol="diamond")
es.add("", [50], [50], symbol_size=16, effect_scale=5.5, effect_period=3,symbol="arrow")
es.add("", [60], [60], symbol_size=13, effect_scale=3.5, effect_period=3,symbol="triangle")
#es.show_config()
es.render(path = "动态散点图各种图形.html")
es
```
