```python
from pyecharts import Line
attr =["", "", "", "", "", ""]
v1 =[5, 20, 36, 10, 10, 100]
v2 =[55, 60, 16, 20, 15, 80]
```

```python
# 
line = Line('')
line.add('A', attr, v1, mark_point=['max'])
line.add('B', attr, v2, mark_point=['min'], is_smooth=True)
#line.show_config()
#line.render('.html')
line
```

```python
# 
line2 = Line('')
line2.add('A', attr, v1,  is_step=True, is_label_show=True)
#line2.show_config()
#line2.render('.html')
line2
```

```python
# 
line3 =Line("")
line3.add("A", attr, v1,  line_opacity=0.2,   area_opacity=0.4, symbol=None, mark_point=['max'])
line3.add("B", attr, v2,  area_color='#a3aed5', area_opacity=0.3, is_smooth=True)
# line3.show_config()
#line3.render('.html')
line3
```

```python
from pyecharts import Pie
attr =["", "", "", "", "", ""]
v1 =[5, 20, 36, 10, 10, 100]
v2 =[55, 60, 16, 20, 15, 80]
```

```python
# 
pie = Pie('')
pie.add('', attr, v1, is_label_show=True)
# pie.show_config()
#pie.render('.html')
pie
```

```python
# 
pie2 = Pie("-", title_pos='center', width=900)
pie2.add("A", attr, v1, center=[25, 50], is_random=True, radius=[30, 75], rosetype='radius',is_label_show=True)
pie2.add("B", attr, v2, center=[75, 50], is_random=True, radius=[30, 75], rosetype='area', is_legend_show=False)
# pie2.show_config()
pie2.render('.html')
pie2
```

```python
from pyecharts import  Scatter
# 
v1 =[10, 20, 30, 40, 50, 60]
v2 =[10, 20, 30, 40, 50, 60]
v3 =[60, 50, 40, 30, 20, 10]
scatter =Scatter("")
scatter.add("A", v1, v2)
scatter.add("B", v1, v3)
#scatter.show_config()
#scatter.render('.html')
scatter
```

```python
from pyecharts import  EffectScatter
attr =["", "", "", "", "", ""]
v1 =[5, 20, 36, 10, 10, 100]
v2 =[55, 60, 16, 20, 15, 80]

# 
es =EffectScatter("")

# v1 x v2 y
es.add("", v1, v2)
#es.show_config()
es.render('.html')
es
```

```python
# 
es = EffectScatter("")
es.add("", [10], [10], symbol_size=20, effect_scale=3.5,  effect_period=3, symbol="pin")
es.add("", [20], [20], symbol_size=12, effect_scale=4.5, effect_period=4,symbol="rect")
es.add("", [30], [30], symbol_size=30, effect_scale=5.5, effect_period=5,symbol="roundRect")
es.add("", [40], [40], symbol_size=10, effect_scale=6.5, effect_brushtype='fill',symbol="diamond")
es.add("", [50], [50], symbol_size=16, effect_scale=5.5, effect_period=3,symbol="arrow")
es.add("", [60], [60], symbol_size=13, effect_scale=3.5, effect_period=3,symbol="triangle")
#es.show_config()
es.render(path = ".html")
es
```
