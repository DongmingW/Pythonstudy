```python
# 2G表格预处理
g2 = pd.read_excel('./2.xls',sheet_name='Sheet1',dtype={'BCCH':str,'挂高':str})
# g3 = pd.read_excel('./3.xls',sheet_name='临沂WCDMA工参')
# g4 = pd.read_excel('./4.xlsx',sheet_name='LTE Project Parameters',skiprows=2)
# netm = pd.DataFrame(columns=['Gen','MCC','MNC','CellID','xAC','NodeBID','PCI','Lon','Lat','Cellname','Freq'])

g2['NCCBCC'] = g2['NCC'].map(str) + g2['BCC'].map(str)
g2['Gen'] = '2G'
g2['MCC'] = 460
g2['MNC'] = 1
g2['cellname'] = g2['bts_name'].map(str) + '方位角' + g2['方位角'].map(str) + '下倾角' + g2['下倾角'].map(str) + '挂高' + g2['挂高'].map(str)
gc2 = g2[['Gen','MCC','MNC','CI','LAC','BTS','NCCBCC','经度','纬度','cellname','BCCH']]
gc2.columns = ['Gen','MCC','MNC','CellID','xAC','NodeBID','PCI','Lon','Lat','Cellname','Freq']
# netm['CellID'] = g2['CI'].append(g3['CellID'],ignore_index=True).append(g4['ECI'],ignore_index=True)
# netm['MCC'] = 460
# netm['CellID'] = g2['CI'].append(g3['CellID'])
# netm.describe
gc2.head()
```

```python
# 3G表格预处理
g3 = pd.read_excel('./3.xls',sheet_name='临沂WCDMA工参',dtype={'DUARFCN':str,'ANTHeight':str})

g3['Gen'] = '3G'
g3['MCC'] = 460
g2['MNC'] = 1
g3['cellname'] = g3['CellName'].map(str) + '方位角' + g3['Azimuth'].map(str) + '下倾角' + g3['MTILT'].map(str) + '挂高' + g3['ANTHeight'].map(str) + g3['塔型'].map(str) + g3['设备类型'].map(str)
gc3 = g3[['Gen','MCC','MNC','CellID','LAC','NodeBID','PSC','CellLON','CellLAT','cellname','DUARFCN']]
gc3.columns = ['Gen','MCC','MNC','CellID','xAC','NodeBID','PCI','Lon','Lat','Cellname','Freq']
gc3.head()
```

```python
# 4G表格预处理
g4 = pd.read_excel('./4.xlsx',sheet_name='LTE Project Parameters',skiprows=2,dtype={'Downlink Center Frequency':str,'ANTHeight':str})
g4['Gen'] = '4G'

g4['下倾角'] = g4['Mechanical Downtilt'].map(str) + '+' +g4['Electrical Downtilt'].map(str)
g4['cellname'] = g4['UserLabel'].map(str) + '方位角' + g4['Azimuth'].map(str) + '下倾角' + g4['下倾角'].map(str) + '挂高' + g4['Height (m)'].map(str) + g4['塔型'].map(str)
gc4 = g4[['Gen','MCC','MNC','Cell ID','TAC','ME ID','PCI','Cell Longitude','Cell Latitude','cellname','Downlink Center Frequency']]
gc4.columns = ['Gen','MCC','MNC','CellID','xAC','NodeBID','PCI','Lon','Lat','Cellname','Freq']
gc4.head()
```

```python
import time
netm = pd.concat([gc2,gc3,gc4],ignore_index=True,axis=0)
# netm.head()
time = time.strftime("%Y%m%d")
outputfilename = './Netmonster%s.csv' % time
outputfilename2 = './Netmonster%s工参.ntm' % time
netm.to_csv(outputfilename,index=0,header=0,encoding='utf-8')
```

```python
f = open(outputfilename,'r',encoding='utf-8')
f_new = open(outputfilename2,'w',encoding='utf-8')
for line in f:
    # 进行判断
    if "," in line:
        line = line.replace(',',';')
    # 如果不符合就正常的将文件中的内容读取并且输出到新文件中
    f_new.write(line)

f.close()
f_new.close()
```
