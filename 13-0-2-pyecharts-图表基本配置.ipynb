## 一、图表配置-图像初始化   
####     图表配置主要是一些参数设置，比如标题、标题位置、标题字体、颜色、画布大小、背景颜色等等
####     以下介绍图像初始化部分，图像初始化的参数一般写在图像函数中
##   1. 标题相关参数
####   


```python
from pyecharts import Bar
bar = Bar(
    title='MainTitle 测试',  #主标题
    subtitle='SubTitle',  #副标题
    title_pos='left',  #主标题位置
    title_color='black',  #标题颜色
    subtitle_color='red'  #副标题颜色
)
#bar.print_echarts_options()
# 图例名称
name = '服装'
# 图例位置
legend_pos = 'center'
# x轴值
x_axis = ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
# y轴值
y_axis = [5, 20, 36, 10, 75, 90]
bar.add(name=name, x_axis=x_axis, y_axis=y_axis, legend_pos=legend_pos)
```

```python
bar.print_echarts_options()
```

## 二、图表设置-通用配置项
#### 通用配置项繁杂多样，更像是小控件的细节设置。主要坐标轴、图例、标签等等～ 
#### 通用配置项均在add()中设置，可以应用于不同的画图函数 ，进行精细化设置。
## 1. 坐标轴
#### 坐标轴的设置比较多，比如坐标轴的颜色、名称、刻度等等
## 1.1坐标轴名称&字体&位置相关参数




```python
from pyecharts import Bar
bar = Bar(
    title='MainTitle',  #主标题
    title_pos='left',  #主标题位置
    title_color='red'  #标题颜色
)
# 图例名称
name = '服装'
# 图例位置
legend_pos = 'center' 
# x轴值
x_axis = ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
# y轴值
y_axis = [5, 20, 36, 10, 75, 90]
# x轴类型,默认情况下应该是自动识别的
xaxis_type = 'category'
# x轴名称
xaxis_name = '种类'
# y轴名称
yaxis_name = '销量'
# x轴名称放在最后
xaxis_name_pos = 'middle'
# y轴名称放在中间
yaxis_name_pos = 'end'
# x轴名称与轴线距离
xaxis_name_gap = 40
bar.add(
    name=name,
    x_axis=x_axis,
    y_axis=y_axis,
    legend_pos=legend_pos,
    xaxis_type=xaxis_type,
    xaxis_name=xaxis_name,
    yaxis_name=yaxis_name,
    xaxis_name_pos=xaxis_name_pos,
    yaxis_name_pos=yaxis_name_pos,
    xaxis_name_gap=40)
```

## 1.2坐标轴显示&方向相关参数



```python
from pyecharts import Bar
bar = Bar(
    title='MainTitle',  #主标题
    title_pos='center',  #主标题位置
    title_color='red'  #标题颜色
)
# 图例名称
name = '服装'
# 图例位置
legend_pos = 'right'
# x轴值
x_axis = ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
# y轴值
y_axis = [5, 20, 36, 10, 75, 90]
# 不显示x轴轴线
is_xaxis_show = True
# 显示y轴轴线
is_yaxis_show =  True
# 显示 y 轴网格线
is_splitline_show = False
# y轴反向
is_yaxis_inverse =  False
# y轴两边不留白
is_xaxis_boundarygap = True
bar.add(
    name=name,
    x_axis=x_axis,
    y_axis=y_axis,
    legend_pos=legend_pos,
    is_xaxis_show=is_xaxis_show,
    is_yaxis_show=is_yaxis_show,
    is_splitline_show=is_splitline_show,
    is_yaxis_inverse=is_yaxis_inverse,
    is_xaxis_boundarygap=is_xaxis_boundarygap)
```

## 1.3 坐标轴标签相关参数


```python
from pyecharts import Bar
bar = Bar(
    title='MainTitle',  #主标题
    title_pos='left',  #主标题位置
    title_color='red'  #标题颜色
)
# 图例名称
name = '服装'
# 图例位置
legend_pos = 'center'
# x轴值
x_axis = ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
# y轴值
y_axis = [5, 20, 36, 10, 75, 90]
# x轴标签与刻度线对齐
is_xaxislabel_align = False
# y轴标签与刻度线不对齐
is_yaxislabel_align = True
# x轴标签字体大小
xaxis_label_textsize = 14
# x轴标签字体颜色
xaxis_label_textcolor = 'red'
# x轴标签与轴线之间的距离
xaxis_margin = 15
bar.add(
    name=name,
    legend_pos=legend_pos,
    x_axis=x_axis,
    y_axis=y_axis,
    is_yaxislabel_align=is_yaxislabel_align,
    xaxis_label_textsize=xaxis_label_textsize,
    xaxis_label_textcolor=xaxis_label_textcolor,
    xaxis_margin=xaxis_margin,
    is_xaxislabel_align=is_xaxislabel_align)
```

## 1.4 坐标轴刻度&间隔



```python
from pyecharts import Bar
bar = Bar(
    title='MainTitle',  #主标题
    title_pos='center',  #主标题位置
    title_color='red'  #标题颜色
)
# 图例名称
name = '服装'
# 图例位置
legend_pos = 'right'
# x轴值
x_axis = ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
# y轴值
y_axis = [10, 20, 36, 10, 75, 90]
# y轴最小值
yaxis_min = 5
# y轴最大值
yaxis_max = 120
# x 轴刻度标签顺时针旋转45度
xaxis_rotate = 45
# x 轴刻度标签的显示间隔，第1，3，5个标签显示
xaxis_interval = 1
# 强制刻度间隔
yaxis_force_interval = 15
bar.add(
    name=name,
    legend_pos=legend_pos,
    x_axis=x_axis,
    y_axis=y_axis,
    yaxis_min=yaxis_min,
    yaxis_max=yaxis_max,
    xaxis_rotate=xaxis_rotate,
    xaxis_interval=xaxis_interval,
    yaxis_force_interval=yaxis_force_interval)
```

## 2 图例 legend常用的参数，如显示的位置，排列的方式等等。


```python
from pyecharts import Bar
bar = Bar(
    title='MainTitle',  #主标题
    title_pos='center',  #主标题位置
    title_color='red'  #标题颜色
)
# 图例名称
name = '服装'
# 图例位置
legend_pos = 'right'
# x轴值
x_axis = ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
# y轴值
y_axis = [10, 20, 36, 10, 75, 90]
# 显示图例
is_legend_show = True
# 图例朝向，一般多个列表时用该参数
legend_orient = 'vertical'
# 图例距离左侧的位置，也就是水平方向位置
legend_pos = 'right'
# 图例距离上侧距离
legend_top = 'top'
# 图例字体颜色
legend_text_color = 'red'
bar.add(
    name=name,
    x_axis=x_axis,
    y_axis=y_axis,
    is_legend_show=is_legend_show,
    legend_orient=legend_orient,
    legend_pos=legend_pos,
    legend_top=legend_top,
    legend_text_color=legend_text_color)
bar.add(
    name='服装2',
    x_axis=["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"],
    y_axis=[10, 25, 8, 60, 20, 80],
    legend_orient=legend_orient,
    legend_pos=legend_pos,
    legend_top=legend_top,
    legend_text_color=legend_text_color
)
```

## 3 文本标签(label):标签显示、位置、字体大小&颜色等等参数设置


```python
from pyecharts import Bar
bar = Bar(
    title='MainTitle',  #主标题
    title_pos='center',  #主标题位置
    title_color='red'  #标题颜色
)
# 图例名称
name = '服装'
# 图例位置
legend_pos = 'right'
# x轴值
x_axis = ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
# y轴值
y_axis = [10000, 20000, 36000, 10000, 75000, 90000]
# 标签颜色
label_color = ['blue']
# 高亮标签位置
label_emphasis_pos = 'right'
is_label_show = True
bar.add(
    name=name,
    legend_pos=legend_pos,
    x_axis=x_axis,
    y_axis=y_axis,
    is_label_show = is_label_show,
    yaxis_label_textcolor='red',
    label_color=label_color,
label_emphasis_pos=label_emphasis_pos)
```

## 4 线的类型(lineStyle)


```python
from pyecharts import Line

attr = ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
v1 = [5, 20, 36, 10, 10, 200]
v2 = [55, 60, 16, 20, 15, 80]
line = Line("折线图示例")
line.add("商家A", attr, v1, mark_point=["average"],line_color='blue',line_width=2,line_type="dotted")
line.add("商家B", attr, v2, is_smooth=True, mark_line=["max", "average"],line_color='red',line_type="dashed")
```
