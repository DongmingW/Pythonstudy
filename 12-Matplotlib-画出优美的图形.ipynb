# Matplotlib-画出优美的图形


Matplotlib 是一个 Python 的 2D绘图库，通过 Matplotlib，可以仅需要几行代码，便可以生成绘图，直方图，功率谱，条形图，错误图，散点图等。

## Matplotlib简介
* Matplotlib是非常强大的python画图工具
* Matplotlib可以画图线图、散点图、等高线图、条形图、柱形图、3D图形、图形动画等。

## Matplotlib基本应用

###  折线图

```python
#可以将matplotlib的图表直接嵌入到Notebook之中 ，使用pycharm或 spyder时注释此句即可
%matplotlib inline   
import matplotlib.pyplot as plt
import numpy as np
 
x = np.arange(0, 10) #定义x
print(x) 
y = x *3 
plt.title("y=x*3")
plt.plot(x,y)#plot()画出曲线
plt.show()#显示图像
```

```python
#在一张画布上画两个图

x=np.arange(-5,5,0.1)#50为生成的样本数
y1=2*x+1
y2=x**2
plt.plot(x,y1,color='red',linewidth=2,linestyle='--')#颜色为红色，线宽度为2，线风格为--
plt.plot(x,y2,color='blue',linewidth=5,linestyle=':')# 颜色为蓝色，线宽带为5，线风格：
plt.xlabel("I'm x")  #定义x轴名称
plt.ylabel("I'm y") #定义y轴名称
plt.show()
```

### 散点图-scatter

```python
np.random.random(100)


X=np.random.random(100)#生成100个1以内的随机数
Y=np.random.random(100)#生成100个1以内的随机数
T=np.around(X+Y) # 设置散点的颜色，将X与Y之和四舍五入  共有三种取值 0 ，1,2
#scatter画散点图  颜色为T   size为100,alpha透明度为60%
plt.scatter(X,Y,s=100,c=T,alpha=0.6 )

plt.show()
```

### 条形图-bar

```python

X = np.arange(6)
Y1=[52,55,63,53,100,33]
Y2=[30,80,100,2,66,77]

bar_width=0.3  # 宽度
plt.bar(X,Y1,bar_width,color='blue') #直方图   
# plt.bar(X,Y2,bar_width,color='red') #直方图   将Y2的数据与Y1合并
plt.bar(X+bar_width,Y2,bar_width,color='red') #直方图   将Y2的数据放在Y1之后

plt.show()
```

此外，还可以绘制饼图、等高线图、3D图等，也可以使用subplot来组合许多的小图在大图中显示。
