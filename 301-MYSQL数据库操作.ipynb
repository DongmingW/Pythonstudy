```python
import pymysql
```

```python
# 打开数据库连接
db = pymysql.connect(host="localhost",port=3333,user="root",passwd="root",db="train")
cursor = db.cursor()

# 使用 execute() 方法执行 SQL，如果表存在则删除
cursor.execute("DROP TABLE IF EXISTS STUDENT1")
 
# 使用预处理语句创建表
sql = """CREATE TABLE `student1` (
        `s_id` INT(11) UNSIGNED NOT NULL AUTO_INCREMENT,
        `name` CHAR(50) NULL DEFAULT NULL,
        `class` CHAR(50) NULL DEFAULT NULL,
        `c_id` INT(11) NULL DEFAULT NULL,
        `score` INT(11) NULL DEFAULT NULL,
        PRIMARY KEY (`s_id`),
        INDEX `Index 2` (`score`)
)
COLLATE='utf8_general_ci'
ENGINE=InnoDB
AUTO_INCREMENT=1007
"""
 
cursor.execute(sql)



# 关闭数据库连接
db.close()
```
