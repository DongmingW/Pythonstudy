```python
# 静态方法和类方法

# 实例方法、静态方法和类方法
# 在实例方法中，可以调用静态方法和类方法，反之，不成立
class MyClass:
    name = '王宁'   # 静态变量
    def __init__(self):
        print('MyClass的构造方法被调用')
        self.value =20
    
    # 定义静态方法
    @staticmethod
    def run():
        print('*',MyClass.name, '*')
        print('MyClass的静态方法run被调用')
    # 定义类方法
    @classmethod
    def do(self):
        print(self)
        print('[',self.name,']')
        self.run()
       
        # print(self.value)
    # 实例方法
    def do1(self):
        print(self.value)
        print('<', self.name, '>')
        print(self)
        self.run()


print("-------------调用静态方法：------------------")
MyClass.run()

print("-------------调用类方法：--------------------")
MyClass.do()

print("-------------调用实例方法：------------------")
c = MyClass()
c.do()
c.do1()
c.run()

```

```python
# 方法重写
class people:
    name = ''    
    age = 0
    __weight = 0
    num = 0
    def __init__(self,n,a,w):
        self.name = n
        self.age = a
        self.__weight = w
    def speak(self):
        print("%s 说: 我 %d 岁。" %(self.name,self.age))
class student(people):
    grade = ''
    def __init__(self,n,a,w,g):
        people.__init__(self,n,a,w)     #调用父类的构函
        self.grade = g
    def speak(self):    #覆写父类的方法
        print("%s 说: 我 %d 岁了，我在读 %s"%(self.name,self.age,self.grade))

p1 = student("王毅",30,50,"三年级")
p1.speak() 
```

```python
# 多态：
class Animal:
    def run(self):
        pass

class People(Animal):
    def run(self):
        print('人正在走')

class Pig(Animal):
    def run(self):
        print('pig is walking')

class Dog(Animal):
    def run(self):
        print('dog is running')

```

```python
# 多态性：

def func(obj):
    obj.run()

peo1=People()
pig1=Pig()
d1=Dog()


func(peo1)
func(pig1)
func(d1)
```

```python
class stack:
    name = ''    
    age = 0
    __weight = 0
    num = 0
    def __init__(self,n,varlist):
        self.n = n
        self.varlist = varlist
        print(self.varlist)
    def pop(self):
        pop1 = self.varlist.pop()
        self.n = self.n - 1
        return pop1
    def push(self,i):
        self.varlist.append(i)
        self.n = self.n + 1
    def len(self):
        n = self.n
        return n
           
sta1 = stack(5,[1,2,3,4,5])
n = sta1.pop()
print("从堆栈中弹出的值为：",n)
sta1.push(6)
print("堆栈值为：",sta1.varlist)
print("堆栈长度为：",sta1.len())
```
